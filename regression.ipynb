## Predicting Sleep Duration from Stress Level, physical activity, and Gender. 

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch import nn as nn
from torch import optim as optim
import pandas as pd
```

Import pre-cleaned data

```python
data = pd.read_csv('clean_data.csv')
print(data.head())
print(data.info())
```

## Split Data

```python
data['Gender'] = data['Gender'].astype('category').cat.codes

features = data[['Gender', 'Sleep.Duration', 'Stress.Level', 'Physical.Activity.Level']]
target = data[['Sleep.Duration']]


n = len(data)
np.random.seed(67)
indices = np.arange(n)

np.random.shuffle(indices)
n_train = int(0.8*n)
train_index = np.arange(0, n_train)
test_index = np.arange(n_train, n)

train_data = data.iloc[train_index]
test_data = data.iloc[test_index]

len(train_data)+ len(test_data) == n

## Normalize training data 
avg_sleep_dur = np.mean(test_data['Sleep.Duration'])
avg_stress_lvl = np.mean(test_data['Stress.Level'])
avg_phy_act_lvl = np.mean(test_data['Physical.Activity.Level'])

```
